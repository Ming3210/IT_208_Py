```python
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, None, 22, 20]
score = [8.5, None, 6.5, 9.0]

df = pd.DataFrame({'name': name, 'age': age, 'score': score})


print("Dữ liệu thiếu kiểm tra bằng isna():")
print(df.isna())
print()


df_fill = df.fillna({'age': 21, 'score': 7.5})
print("DataFrame sau khi thay thế NaN bằng fillna():")
print(df_fill)
print()


df_drop = df.dropna()
print("DataFrame sau khi loại bỏ NaN bằng dropna():")
print(df_drop)
```

```python

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({'Name': name, 'Age': age, 'Score': score})

df["Age_Group"] = df["Age"].apply(lambda x: "Young" if x < 21 else "Adult")

mapping = {8.5: "Excellent", 7.0: "Good", 6.5: "Average", 9.0: "Excellent"}
df["Score_Level"] = df["Score"].map(mapping)

df["Name"] = df["Name"].replace("Bình", "Bình (Updated)")

print(df)
```

```python

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({'Name': name, 'Age': age, 'Score': score})

df["Name"] = df["Name"].replace("Bình", "Bình (Updated)")

grouped = df.groupby("Age")["Score"].agg(
    mean_score="mean",
    min_score="min",
    max_score="max"
).reset_index()

print(grouped)
print(df)
```

```python
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['An', 'Bình', 'Châu']})
df2 = pd.DataFrame({'id': [1, 2, 3], 'score': [8.5, 7.0, 6.5]})

merged_df = pd.merge(df1, df2, on="id")
print("Merged DataFrame:")
print(merged_df)

df3 = pd.DataFrame({'age': [20, 21, 22]})

concatenated_df = pd.concat([merged_df, df3], axis=1)
print("\nConcatenated DataFrame:")
print(concatenated_df)
```

```python
import json
import os
from pathlib import Path

os.makedirs('logs', exist_ok=True)

# Tạo file log mẫu
for date, user, action, duration in [
    ('2024-01-01', 'A', 'login', 5),
    ('2024-01-02', 'B', 'logout', 3)
]:
    with open(f'logs/log_{date}.json', 'w') as f:
        json.dump({"user": user, "action": action, "duration": duration}, f)

# Đọc và phân tích
data = []
for file in sorted(Path('logs').glob('log_*.json')):
    date = pd.to_datetime(file.stem.replace('log_', '')).date()
    with open(file) as f:
        log = json.load(f)
        log['date'] = date
        data.append(log)

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

# Tính tổng duration theo ngày
result = df.groupby(df['date'].dt.date)['duration'].sum().reset_index()
result.columns = ['Date', 'Total Duration']

# In kết quả
for _, row in result.iterrows():
    print(f"Date: {row['Date']}, Total Duration: {row['Total Duration']}")

# Xuất CSV
result.to_csv('daily_report.csv', index=False)
print("\n✓ Xuất ra: daily_report.csv")
```
