```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Ct 1: Lineplot
axes[0, 0].plot(x, x)
axes[0, 0].set_title('y = x')

axes[1, 0].plot(x, x**2)
axes[1, 0].set_title('y = x')

# Ct 2: Scatterplot
axes[0, 1].scatter(np.random.rand(50), np.random.rand(50))
axes[0, 1].set_title('Scatter 1')

axes[1, 1].scatter(np.random.rand(50), np.random.rand(50))
axes[1, 1].set_title('Scatter 2')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# To d liu thi gian 30 ngy
dates = pd.date_range("2024-01-01", periods=30)
temp = 20 + np.random.randn(30) * 2  # nhit  ngu nhin

# Trendline (hi quy tuyn tnh)
x = np.arange(30)
coef = np.polyfit(x, temp, 1)
trend = coef[0]*x + coef[1]

# Spike: im tng/gim bt thng
spike_idx = np.where(np.abs(temp - np.mean(temp)) > 2.5)[0]

# Ngy c bit (v d: ngy 15)
special_day = 14

plt.figure(figsize=(10,4))
plt.plot(dates, temp, label="Nhit ", marker="o")
plt.plot(dates, trend, label="Trendline", linestyle="--")

# nh du spike
plt.scatter(dates[spike_idx], temp[spike_idx], color="red", label="Spike")

# nh du ngy c bit
plt.axvline(dates[special_day], color="green", linestyle=":", label="Ngy c bit")

plt.xlabel("Ngy")
plt.ylabel("Nhit  (C)")
plt.title("Phn tch chui thi gian (n gin)")
plt.legend()
plt.tight_layout()
plt.show()
```
