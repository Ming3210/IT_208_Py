```python
def difference(a, b):
    return a - b


def multiply(a, b):
    return a * b

print(difference(10, 5))


print(multiply(10, 5))
```

```python
def isPrime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


print(isPrime(11))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
print(factorial(5))
```

```python
def circle_area(radius):
    import math
    return math.pi * radius ** 2
print(circle_area(5))
```

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True 
    else:
        return False
    
print(is_leap_year(2020))
```

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n = n // 10
    return total


print(sum_of_digits(12345))
```

```python
def handleReverseTheNumber(n):
    reversed_number = 0
    while n > 0:
        digit = n % 10
        reversed_number = reversed_number * 10 + digit
        n = n // 10
    return reversed_number

print(handleReverseTheNumber(12345))
```

```python
def is_perfect_number(n):
    if n < 1:
        return False
    sum_of_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_of_divisors += i
    return sum_of_divisors == n

print(is_perfect_number(6))
```
