```python
import csv

with open('student.csv', mode='r', encoding='utf-8') as src, \
     open('student_new.csv', mode='w', encoding='utf-8', newline='') as dst:

    reader = csv.reader(src)
    writer = csv.writer(dst)

    for row in reader:
        if row[2] == 'grade' or int(row[2]) > 7:
            writer.writerow(row)
            print(row)

```

```python
import json

userList = []

with open('userdata.json', mode='r', encoding='utf-8') as f:
    userList = json.load(f)
    for user in userList:
        if user["id"] == 3:
            print(user)

newUser = {
    "id": 5,
    "name": "JohnSon",
    "age": 30
}

userList.append(newUser)

with open('userdata.json', mode='w', encoding='utf-8') as f:
    json.dump(userList, f,  indent=4)
```

```python
import json

inventoryList = []

with open('inventory.json', mode='r', encoding='utf-8') as f:
    inventoryList = json.load(f)


newInventory = {
    "product_id": 4,
    "name": "Potions",
    "quantity": 50,
    "price": 50
}

inventoryList.append(newInventory)
    
with open('inventory.json', mode='w', encoding='utf-8') as f:
    json.dump(inventoryList, f,  indent=4)
```
