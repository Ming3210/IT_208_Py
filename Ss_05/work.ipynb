```python
import csv

with open('student.csv', mode='r', encoding='utf-8') as src, \
     open('student_new.csv', mode='w', encoding='utf-8', newline='') as dst:

    reader = csv.reader(src)
    writer = csv.writer(dst)

    for row in reader:
        if row[2] == 'grade' or int(row[2]) > 7:
            writer.writerow(row)
            print(row)

```

```python
import json

userList = []

with open('userdata.json', mode='r', encoding='utf-8') as f:
    userList = json.load(f)
    for user in userList:
        if user["id"] == 3:
            print(user)

newUser = {
    "id": 5,
    "name": "JohnSon",
    "age": 30
}

userList.append(newUser)

with open('userdata.json', mode='w', encoding='utf-8') as f:
    json.dump(userList, f,  indent=4)
```

```python
from datetime import datetime

today_str = datetime.now().strftime("%Y-%m-%d")

with open("log.txt", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        
        timestamp_str = line[:19]
        
        try:
            timestamp = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
        except ValueError:
            continue
        
        if timestamp.strftime("%Y-%m-%d") == today_str:
            print(line)
```
