```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")

arr = np.random.normal(7, 1.5, 200) 
arr = np.clip(arr, 0, 10)
plt.figure(figsize=(10, 5))
sns.histplot(arr, bins=20, kde=True, color="green")  

plt.title("Biểu đồ phân phối điểm thi")
plt.xlabel("Điểm")
plt.ylabel("Tần suất")
plt.show()

# Phân phối tương đối cân đối, đa số điểm nằm 6-8
```

```python
import pandas as pd
subjects = ["Math","Math", "Math", "Physics", "Physics", "Physics", "Chemistry","Chemistry"]
scores = [8,7,7,9,9,10,8,9]
df = pd.DataFrame({"subjects": subjects, "scores": scores})
#Phân phối lệch phải, đa số điểm nằm 6–8

sns.displot(df, x="scores", hue="subjects", kind="kde")
```
