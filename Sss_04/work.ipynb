```python
products = {
    "A": 120,
    "B": 80,
    "C": 300,
    "D": 50
}

for key, value in products.items():
    if key == "A":
        print(key, value)
        break
    else:
        print("Not found")

products_sorted = dict(sorted(products.items(), key=lambda item: item[1], reverse=False))
newProduct = {}
print(products_sorted)

total = sum(products.values())

print("Total price:", total)

for key, value in products.items():
    if value <= 200:
        newProduct[key] = value


print(newProduct)


```
