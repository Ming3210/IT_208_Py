```python
v5 = input("Nhập một số nguyên: ")

for i in range((len(v5)-1)//2, -1, -1):
    if v5[i] != v5[len(v5)-1-i]:
        print("Không phải số đối xứng")
        break
else:
    print("Là số đối xứng")
```

```python
v6 = int(input("Nhập một số nguyên: "))
result = 1
if (v6 < 0):
    print("Số vừa nhập là số âm")
else:
    for i in range(1, v6 + 1):
        result = result * i
print(result)
```

```python
def isPrime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


a = int(input("Nhập một số nguyên a: "))
b = int(input("Nhập một số nguyên b: "))
if a > b:
    print("a lớn hơn b")

for i in range(a, b + 1, 1):
    if isPrime(i):
        print(i)
```

```python
def isTriangle(a, b, c):
    if a + b > c and a + c > b and b + c > a:
        return True
    return False


print(isTriangle(3, 4, 5))
```
