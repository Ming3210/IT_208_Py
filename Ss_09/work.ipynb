```python
import matplotlib.pyplot as plt

import numpy as np

fig, ax = plt.subplots(figsize=(8, 6))

x = np.random.randint(1, 10, 10)
y = x**2

ax.grid()
ax.set_title("Biu  bng")
ax.plot(x, y)

plt.show()
```

```python


x = np.arange(-5, 6)   
y = x**3

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, color="red", linewidth=3)  

ax.set_title(" th hm bc ba y = x")
ax.set_xlabel("Gi tr x")
ax.set_ylabel("Gi tr y")

ax.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

```python
x = np.arange(0,11)
y = 2*x+3

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, color="green", linestyle="--", linewidth=3)

ax.set_title(" th hm bc ba y = 2x + 3")
ax.set_xlabel("Gi tr x")
ax.set_ylabel("Gi tr y")

ax.grid(True, linestyle="--")
```

```python
x = np.linspace(0, 2*3.14)
y = np.sin(x)
z = np.cos(x)


fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, color="cyan", linewidth=3)

ax.plot(x, z, color="red",linestyle = "--", linewidth=3)

ax.set_xlabel("Gi tr x")
ax.set_ylabel("Gi tr y")

```

```python
x = np.linspace(0,11)

y1 = 2*x+1
y2 = -x+10

fig,ax = plt.subplots(figsize = (8,6))

ax.plot(x, y1, color="cyan", linewidth=3)

ax.plot(x, y2, color="red",linestyle = "--", linewidth=3)

```

```python
years = list(range(2015, 2025))
population = [92, 93, 94, 95, 96, 97, 98, 99, 100, 101]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(years, population, color="blue", linestyle="-", marker="o", linewidth=2, markersize=6, label="Dn s")
ax.set_title("Tng trng dn s Vit Nam 20152024")
ax.set_xlabel("Nm")
ax.set_ylabel("Dn s (triu ngi)")
ax.set_xticks(years)
ax.grid(True)
ax.legend()
plt.show()
```

```python
products = ["o", "Qun", "Giy", "Ti", "M", "Tt"]
sales = [150, 120, 200, 90, 65, 80]

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(products, sales, color="green", edgecolor="black")
ax.set_ylim(0, max(sales) * 1.12)
for bar in bars:
    h = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        h * 0.98,              
        f"{int(h)}",
        ha="center",
        va="top",
        color="white",
        fontweight="bold"
    )

ax.set_title("Doanh s bn hng thng 10/2025")
ax.set_ylabel("Doanh s (triu)")
ax.set_xlabel("Sn phm")
ax.grid(False)

plt.show()
```

```python


months = list(range(1, 13))
temp = [20, 22, 25, 28, 30, 32, 31, 30, 28, 26, 23, 21]          # Nhit  (C)
rain = [10, 15, 30, 120, 200, 250, 280, 220, 150, 80, 30, 15]    # Lng ma (mm)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(months, temp, color="red", linestyle="-", linewidth=2, marker="o", markersize=6, label="Nhit  (C)")

ax2 = ax.twinx()
bars = ax2.bar(months, rain, color="blue", alpha=0.7, width=0.6, label="Lng ma (mm)")

ax.set_xlabel("Thng")
ax.set_ylabel("Nhit  (C)", color="red")
ax2.set_ylabel("Lng ma (mm)", color="blue")
ax.set_xticks(months)

handles1, labels1 = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax.legend(handles1 + handles2, labels1 + labels2, loc="upper left")

ax.grid(True)
ax.set_title("Nhit  v lng ma H Ni 2025")

plt.show()
#a
```
