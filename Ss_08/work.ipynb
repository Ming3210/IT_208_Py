```python

students = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["An", "Binh", "Chau"]
})

scores = pd.DataFrame({
    "id": [1, 1, 2, 3, 3, 3],
    "score": [8.5, 9.0, 7.5, 6.0, 7.0, 8.0]
})

merged = students.merge(scores, on="id")

report = merged.groupby(["id", "name"])["score"].agg(["min","max","mean"]).reset_index()

report.to_csv("report.csv", index=False)
print(report)
```

```python
import pandas as pd
import numpy as np

# Tạo dataset
data = {
    'name': ['An', 'Bình', 'Châu', 'Dũng'],
    'age': [20, np.nan, 22, 21],
    'score': [8.5, np.nan, 7.0, 9.0],
    'city': ['Hà Nội', 'HCM', np.nan, 'Hà Nội']
}

df = pd.DataFrame(data)

print("Dữ liệu thiếu:\n", df.isna())

df['age'] = df['age'].fillna(df['age'].mean())     
df['score'] = df['score'].fillna(df['score'].median())  
df['city'] = df['city'].fillna('Unknown')            
df['city'] = df['city'].astype('category')

def score_level(score):
    if score >= 8:
        return 'Good'
    elif score >= 7:
        return 'Average'
    else:
        return 'Bad'

df['score_level'] = df['score'].apply(score_level)

age_mean = df['age'].mean()
age_std = df['age'].std()
df['age_normalized'] = (df['age'] - age_mean) / age_std


age_cleaned = df['age'].tolist()
score_cleaned = df['score'].tolist()
city_category = df['city']
score_level_list = df['score_level'].tolist()
age_normalized = df['age_normalized'].tolist()

print("\nage_cleaned =", age_cleaned)
print("score_cleaned =", score_cleaned)
print("city (category) =", city_category)
print("score_level =", score_level_list)
print("age_normalized =", age_normalized)
```

```python
import os
from pathlib import Path

os.makedirs('logs', exist_ok=True)

# Tạo file log mẫu
for date, user, action, duration in [
    ('2024-01-01', 'A', 'login', 5),
    ('2024-01-02', 'B', 'logout', 3)
]:
    with open(f'logs/log_{date}.json', 'w') as f:
        json.dump({"user": user, "action": action, "duration": duration}, f)

# Đọc và phân tích
data = []
for file in sorted(Path('logs').glob('log_*.json')):
    date = pd.to_datetime(file.stem.replace('log_', '')).date()
    with open(file) as f:
        log = json.load(f)
        log['date'] = date
        data.append(log)

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])

# Tính tổng duration theo ngày
result = df.groupby(df['date'].dt.date)['duration'].sum().reset_index()
result.columns = ['Date', 'Total Duration']

# In kết quả
for _, row in result.iterrows():
    print(f"Date: {row['Date']}, Total Duration: {row['Total Duration']}")

# Xuất CSV
result.to_csv('daily_report.csv', index=False)
print("\n✓ Xuất ra: daily_report.csv")
```
